```python
import cv2 as cv #thư viện opencv để xử lý ảnh thô chuyển thành dữ liệu
import numpy as np # thư viện numpy
import os #lấy đường dẫn tới thư mục chứa hình ảnh
```

###  tạo data

```python
X = []#tạo list chứa dữ liệu ảnh
y = []#tạo list chứa label của từng ảnh
labels = os.listdir('data/train')#gọi biến labels là list gồm các thư mục data con chứa ảnh 

#tạo đường dẫn vào vào dataset
path='data'
dataset = 'train'

#cho biến label trong vòng lặp list labels 
for label in labels:
        
        # Và lặp các ảnh trong thư mục
        for file in os.listdir(os.path.join(path, dataset, label)):
            # đọc ảnh            
            image = cv.imread(os.path.join(path, dataset, label, file),cv.IMREAD_UNCHANGED)
            #chuyển ảnh cấu trúc màu BGR thành RGB
            image =cv.cvtColor(image,cv.COLOR_BGR2RGB)
            image =cv.GaussianBlur(image,(1,1),0,0)
            # chuyển tên các label thành dạng số 
            classnum = labels.index(label)
            #đưa shape ảnh về kích thước 50x50
            img = cv.resize(image,(200,200))
            img = cv.normalize(img, None, alpha=0,beta=255, norm_type=cv.NORM_MINMAX)

            #đưa dữ liệu ảnh đc xử lý vào list X          
            X.append(img)
            #đưa dữ liệu label vào list y
            y.append(classnum)#đưa list X về dữ liệu dạng mảng tensor

#đưa list X về dữ liệu dạng mảng tensor
X= np.array(X)
#đưa list y vễ dạng mảng
y= np.array(y)
from sklearn.utils import shuffle

#xào trộn dữ liệu

X_train, y_train = shuffle(X, y, random_state=42)
print(X_train.shape)
print(y_train.shape)
```

```python
X_test = []#tạo list chứa dữ liệu ảnh test
y_test = []#tạo list chứa label của từng ảnh test
labels = os.listdir('data/Test')#gọi biến labels là list gồm các thư mục data con chứa ảnh 

#tạo đường dẫn vào vào dataset
path='data'
dataset = 'Test'

#cho biến label trong vòng lặp list labels 
for label in labels:
        
        # Và lặp các ảnh trong thư mục
        for file in os.listdir(os.path.join(path, dataset, label)):
            # đọc ảnh            
            image = cv.imread(os.path.join(path, dataset, label, file),cv.IMREAD_UNCHANGED)
            #chuyển ảnh cấu trúc màu BGR thành RGB
            image =cv.cvtColor(image,cv.COLOR_BGR2RGB)
            image =cv.GaussianBlur(image,(1,1),0,0)
            # chuyển tên các label thành dạng số 
            classnum = labels.index(label)
            #đưa shape ảnh về kích thước 50x50
            img = cv.resize(image,(200,200))
            img = cv.normalize(img, None, alpha=0,beta=255, norm_type=cv.NORM_MINMAX)

            #đưa dữ liệu ảnh đc xử lý vào list X          
            X_test.append(img)
            #đưa dữ liệu label vào list y
            y_test.append(classnum)#đưa list X về dữ liệu dạng mảng tensor

            
from sklearn.utils import shuffle
      
#đưa list X về dữ liệu dạng mảng tensor
X_test= np.array(X_test)
#đưa list y vễ dạng mảng
y_test= np.array(y_test)
#xào trộn dữ liệu
X_test, y_test = shuffle(X_test, y_test, random_state=42)      
print(X_test.shape)
print(y_test.shape)
```

```python
from tensorflow.keras.utils import to_categorical
#onehot label y 
y_train= to_categorical(y_train,10)
y_test= to_categorical(y_test,10)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### train model

```python
#chạy GPU tren jupyter
import tensorflow as tf

gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
    except RuntimeError as e:
           print(e)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten,BatchNormalization
model= Sequential()
model.add(Conv2D(32,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same',input_shape=(200,200,3)))
model.add(Conv2D(64,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D((2,2),strides= 5))

model.add(Conv2D(64,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(Conv2D(128,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D((2,2),strides= 5))

model.add(BatchNormalization())


model.add(Conv2D(128,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(Conv2D(512,(3,3),activation='relu',kernel_initializer='he_uniform',padding='same'))
model.add(MaxPooling2D((2,2),strides= 5))

model.add(BatchNormalization())

model.add(Flatten())
model.add(Dense(128,activation='relu',kernel_initializer='he_uniform'))
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam,SGD
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
#opt= SGD(lr=0.01,momentum=0.9)
#sử dụng optimizer Adam
opt =Adam()
#cấu hình train
model.compile(opt,loss='categorical_crossentropy',metrics=['accuracy'])
#bỏ dữ liệu vô train
history = model.fit(X_train, y_train, epochs=20, batch_size=64,validation_data=(X_test,y_test),callbacks=[callback])
#lưu model
model.save('contrung1.h5')
     
```

### tạo GUI

```python
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import tensorflow as tf
import numpy as np
#tạo label
labels = ['kiến','ong','bọ','bướm','sâurom','gián']
# load model
model = tf.keras.models.load_model('contrung1.h5')

# tạo window GUI
window = tk.Tk()
#tiêu đề cho window
window.title("Image Prediction")

# hàm xử lý và nhận diện ảnh
def predict_image():
    # mở file thư mục để chọn ảnh
    file_path = filedialog.askopenfilename()
    
    # load và xử lý ảnh

    image = cv2.imread(file_path)
    image =cv.cvtColor(image.copy(),cv.COLOR_BGR2RGB)
    image =cv.GaussianBlur(image,(3,3),0,0)
    image = cv.resize(image,(200,200))
    image = cv.normalize(image, None, alpha=0,beta=255, norm_type=cv.NORM_MINMAX)
    
    # đưa vô model để dự đoán kết quả
    prediction = model.predict(image.reshape(1,200,200,3))
    
    # đưa kết quả dạng mã hóa về kiểu số nguyên
    class_label = np.argmax(prediction)
    #đưa vô labels từng con côn trùng
    class_label = labels[class_label]
    # hiển thị ảnh được chọn
    image_display.image = ImageTk.PhotoImage(Image.open(file_path))
    image_display.configure(image=image_display.image)
    #hiển thị kế quả
    result_label.configure(text="Predicted Class: {}".format(class_label))

# tạo nút nhấn để kích hoạt hàm xử lý và dự đoán ảnh
upload_button = tk.Button(window, text="Upload Image", command=predict_image)
upload_button.pack(pady=10)

image_display = tk.Label(window)
image_display.pack()

# hiển thị label lên window app
result_label = tk.Label(window, text="")
result_label.pack(pady=10)

# chạy vòn lặp
window.mainloop()
```
